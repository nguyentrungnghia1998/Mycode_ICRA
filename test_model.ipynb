```python
def calculate_total_audio_hours_from_tsv(tsv_file_path, sample_rate=16000):
    total_samples = 0

    # Đọc file .tsv
    with open(tsv_file_path, 'r') as f:
        lines = f.readlines()
    
    # Bỏ qua dòng đầu tiên (đường dẫn thư mục)
    for line in lines[1:]:
        # Tách các cột theo tab, cột thứ hai là số mẫu
        fields = line.strip().split('\t')
        if len(fields) > 1:
            total_samples += int(fields[1])
    
    total_hours = total_samples / 3600000

    return total_hours

# Ví dụ sử dụng
tsv_file_path = "data_mau/public_data/VSV-1000_wav/train.tsv"
total_hours = calculate_total_audio_hours_from_tsv(tsv_file_path)
print(f"Tổng số giờ audio: {total_hours:.2f} giờ")
```
